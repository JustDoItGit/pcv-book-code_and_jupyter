## 1.2.1 绘制图像 、点和线

```python
from PIL import Image
from pylab import array
from pylab import imshow
from pylab import plt
from pylab import plot
from pylab import title
from pylab import hist
from pylab import axis
```

```python
# 读取图像到数组中
im = array(Image.open('../../data/empire.jpg'))
```

```python
import numpy as np
```

```python
im_np = np.array(Image.open('../../data/empire.jpg'))
```

```python
im_np.shape
```

```python
im.shape
```

```python
# 绘制图像
plt.figure(figsize=(12, 6))
imshow(im)
# 一些点
x = [100, 100, 400, 400]
y = [200, 500, 200, 500]
# 使用红色星状标记绘制点
plot(x, y, 'r*')
# 绘制连接前两个点的线
plot(x[:2], y[:2], 'g-')
# 添加标题，显示绘图的图像
title('Plotting: "empire.jpg"')
axis('off')
```

## 1.2.2 图像轮廓和直方图

```python
# 读取图像到数组中
im = np.array(Image.open('../../data/empire.jpg').convert('L'))
```

```python
im.shape
```

```python
im_np[0][0]
```

```python
im[0][0]
```

```python
# 新建一个图像
plt.figure(figsize=(6,9))
#  不使用颜色信息
plt.gray()
# 在原点的左上角显示轮廓图像
plt.contour(im, origin='image')
axis('equal')
axis('off')
plt.show()
```

```python
plt.figsize = (9, 12)
hist(im.flatten(), 128)
plt.show()
```
